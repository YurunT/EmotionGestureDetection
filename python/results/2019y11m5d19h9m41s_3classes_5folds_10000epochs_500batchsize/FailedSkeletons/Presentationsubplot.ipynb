```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
import re
```

```python
repl_str = re.compile(r'[-]?\d[\d,]*[\.]?[\d{2}]*')
pngs = dict()# its length shouble be 30 because of 30 frames
for file in glob.glob("fold0/sec17_label2_into1/*.png"):
    output = []
    match = re.findall(r'\d+', file)[-1]
    pngs[int(match)]=file
```

```python
fig=plt.figure(figsize=(80,80))
for i in range(9):# [0,8]
    img = mpimg.imread(pngs[i])
    plt.subplot(3,3,i+1)
    plt.title("frame"+str(i),fontsize=80)
    plt.imshow(img)
fig.savefig('2to1/2to1label_0to8frame.png')

fig2=plt.figure(figsize=(80,80))
for i in range(9,18):# [9,17]
    img = mpimg.imread(pngs[i])
    plt.subplot(3,3,i+1-9)
    plt.title("frame"+str(i),fontsize=80)
    plt.imshow(img)
fig2.savefig('2to1/2to1label_9to17frame.png')

fig3=plt.figure(figsize=(80,80))
for i in range(18,27):# [18,26]
    img = mpimg.imread(pngs[i])
    plt.subplot(3,3,i+1-18)
    plt.title("frame"+str(i),fontsize=80)
    plt.imshow(img)
fig3.savefig('2to1/2to1label_18to26frame.png')

fig4=plt.figure(figsize=(80,80))
for i in range(27,30):# [27,29]
    img = mpimg.imread(pngs[i])
    plt.subplot(3,3,i+1-27)
    plt.title("frame"+str(i),fontsize=80)
    plt.imshow(img)
fig4.savefig('2to1/2to1label_27to29frame.png')
```
